```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设损失数据存储在 'loss.csv' 文件中
df = pd.read_csv('./rewards.csv')  # 读取数据文件
# df = pd.read_csv('./out/PPA_1/MARDSE-rewards.csv')  # 读取数据文件

# 假设文件中有一列叫 'loss' 存储损失值
reward = df['reward']

# 计算移动平均，设置 min_periods=1 以保留前面的数据点
window_size = 20
reward_smooth = reward.rolling(window=window_size, min_periods=1).mean()

# 绘制损失值的变化图
plt.plot(reward_smooth)
# plt.title('smooth_reward over Iterations')
plt.xlabel('Iterations')
plt.ylabel('reward')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件并返回相应的滚动平均
def read_and_process(file_path, window_size=20):
    data = pd.read_csv(file_path)  # 读取数据文件
    reward = data['reward']        # 提取奖励列
    return reward.rolling(window=window_size, min_periods=1).mean()

# 文件路径列表
file_paths = [
    './out/P_1/MARDSE-rewards.csv', 
    './out/P_1/IPPO-rewards.csv', 
    './out/P_1/PPO-rewards.csv', 
    './out/P_1/TDM-rewards.csv'
]

# 标签对应的名称
labels = ['MARDSE', 'IPPO', 'PPO', 'TDM']

# 颜色对应的配置
colors = ['purple', 'blue', 'cyan', 'red']

# 绘制所有数据集
plt.figure(figsize=(10, 6))  # 可以设置图像大小

for file_path, label, color in zip(file_paths, labels, colors):
    smoothed_reward = read_and_process(file_path)
    plt.plot(smoothed_reward, label=label, color=color, alpha=0.7)

plt.xlabel('Iterations')
plt.ylabel('Reward')
plt.legend()
plt.show()
```
